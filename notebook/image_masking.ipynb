# Image Masking

Image masking is a technique in computer vision and image processing that involves selectively applying a mask (also known as a binary image or a binary mask) to an image to control which parts of the image are visible and which are hidden or modified. The mask is typically a binary image, where each pixel is either set to 1 (white) to keep the corresponding pixel in the original image or set to 0 (black) to hide or modify the corresponding pixel.

## Bitwise Operations and Image Masking with OpenCV

Bitwise operations are used in image processing to extract a Region of Interest (ROI) from an image.

### Bitwise AND

The cv2.bitwise_and function calculates the per-element bit-wise conjunction of two arrays. 

### Bitwise OR

The cv2.bitwise_or function calculates the per-element bit-wise disjunction of two arrays.

### Bitwise XOR

The cv2.bitwise_xor function calculates the per-element bit-wise "exclusive or" operation on two arrays.

### Bitwise NOT

The cv2.bitwise_not function calculates the per-element bit-wise inversion of the input array.




```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# draw a vertical rectangle
image1 = np.zeros((400, 600), dtype="uint8")
cv2.rectangle(image1, (50, 50), (250, 350), 255, -1)

# draw a horizontal rectangle
image2 = np.zeros((400, 600), dtype="uint8")
cv2.rectangle(image2, (50, 200), (550, 350), 255, -1)

# bitwise and mapping
bitwise_and = cv2.bitwise_and(image1, image2)

# bitwise or mapping
bitwise_or = cv2.bitwise_or(image1, image2)

# bitwise xor mapping
bitwise_xor = cv2.bitwise_xor(image1, image2)

# bitwise not
bitwise_not = cv2.bitwise_not(image1)


```

```python
f, axes = plt.subplots(1, 6 , figsize=(16,8))
axes[0].imshow(image1, cmap ='gray')
axes[1].imshow(image2,cmap ='gray')
axes[2].imshow(bitwise_and, cmap = 'gray')
axes[3].imshow(bitwise_or, cmap = 'gray')
axes[4].imshow(bitwise_xor, cmap = 'gray')
axes[5].imshow(bitwise_or, cmap = 'gray')

plt.setp(axes[0], xlabel= "First Image")
plt.setp(axes[1], xlabel= "Second Image")
plt.setp(axes[2], xlabel= "Bitwise AND Mask Image")
plt.setp(axes[3], xlabel= "Bitwise OR Mask Image")
plt.setp(axes[4], xlabel= "Bitwise XOR Mask Image")
plt.setp(axes[5], xlabel= "Bitwise NOT Mask Image")


f.tight_layout()
plt.show()
```

### Masking Image


```python
# Load images 
image_path = '../dataset/moon.jpg'
moon_image = cv2.imread(image_path)
```

```python
plt.imshow(cv2.cvtColor(moon_image, cv2.COLOR_BGR2RGB))
plt.title("Original Images")
```

```python
# create a black image with the same size as our 
# image that contains the moon, we then create
# a white circle on the black image
mask = np.zeros(moon_image.shape[:2], dtype="uint8")
cv2.circle(mask, (202, 133), 34, 255, -1)

# apply the mask to our image
masked = cv2.bitwise_and(moon_image, moon_image, mask=mask)
```

```python
f, axes = plt.subplots(1, 3, figsize=(16,8))
axes[0].imshow(cv2.cvtColor(moon_image, cv2.COLOR_BGR2RGB))
axes[1].imshow(mask,cmap ='gray')
axes[2].imshow(cv2.cvtColor(masked, cv2.COLOR_BGR2RGB))

plt.setp(axes[0], xlabel= "Original Image")
plt.setp(axes[1], xlabel= " Mask ")
plt.setp(axes[2], xlabel= "Masked Image of Moon")



f.tight_layout()
plt.show()
```

- If a mask is applied to a grayscale image of the same size, all the black part of mask are set to zero in the output image. All others remain unchanged.
